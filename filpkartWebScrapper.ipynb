```python
def dataframe(container):
    data = {"Book_Name":[],"price":[]}
    for co in container:
        
        #book name
        title = co.find('div',attrs={'class':'_4rR01T'}).get_text()
        t = title.strip()
        data["Book_Name"].append(t)
       
        #price
        price = co.find('div',attrs={'class':'_30jeq3 _1_WHN1'}).get_text()
        p = price.strip()
        data["price"].append(p)
        
        
        
    return data
            
```

```python
dataset = dataframe(container)
dataset
```

```python
df = pd.DataFrame(dataset)
```
