```python
def dataframe(container):
    data = {"Book_Name":[],"price":[],"date":[]}
    for co in container:
        
        #book name
        title = co.find('span',attrs={'class':'a-size-medium a-color-base a-text-normal'}).get_text()
        t = title.strip()
        data["Book_Name"].append(t)
       
        #price
        price = co.find('span',attrs={'class':'a-offscreen'}).get_text()
        p = price.strip()
        data["price"].append(p)
        
        #date
        date = co.find('span',attrs={'class':'a-size-base a-color-secondary a-text-normal'}).get_text()
        d = date.strip()
        data["date"].append(d)
        
    return data
            
```

```python
dataset = dataframe(container)
dataset
```

```python
df = pd.DataFrame(dataset)
```
